# 행/열 선택법

## 1. 행번호(row number)로 선택하는 방법 (.iloc)

```python
# .iloc
# .iloc[행,열]
df.iloc[0] # 0번
```

```python
df.iloc[0:10,1:4]
# 0~10번 행 + 1~4번 열
```

## 2. label이나 조건표현으로 선택하는 방법 (.loc)

```python
# .loc
# .loc[행 조건 또는 label , 열 조건 또는 label]
df.loc[:,['Name','Pclass']]
```

```python
df.loc[df['Survived'] == True] # Survived == True 인 행 출력
```
