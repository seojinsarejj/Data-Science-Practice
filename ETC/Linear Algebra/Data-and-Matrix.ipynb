## 데이터의 유형

- 스칼라(scalar) : 숫자 하나로 이루어진 데이터
- 벡터(vector) : 여러 숫자로 이루어진 데이터 레코드
- 행렬(matrix) : 데이터 레코드가 여럿인 데이터 집합
- 텐서(tensor) : 같은 크기의 행렬이 여러 개 있는 것

```python
from sklearn.datasets import load_iris

iris = load_iris() # 붓꽃 데이터 로드
iris.data[0,:] # 첫번째 꽃의 데이터
```
