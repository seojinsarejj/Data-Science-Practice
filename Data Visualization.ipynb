```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Line Charts

```python
# Path of the file to read
spotify_filepath = "../data/Visualization_data/spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
spotify_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")


sns.lineplot(data=spotify_data)
```

```python
# Set size
plt.figure(figsize=(14,6))

# Set title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

sns.lineplot(data=spotify_data['Shape of You'],label='Shape of You')
sns.lineplot(data=spotify_data['Despacito'],label='Despacito')

plt.xlabel('Date')
```

# Bar Charts and Heatmaps

```python
# Path of the file to read
flight_filepath = "../data/Visualization_data/flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data
```

## Bar Chart

```python
# Set Size
plt.figure(figsize=(10,6))

# Set Title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

sns.barplot(x=flight_data.index, y=flight_data.NK)

plt.ylabel("Arrival delay (in minutes)")
```

## Heatmap

```python
# Set Size
plt.figure(figsize=(14,7))

# Set Title
plt.title("Average Arrival Delay for Each Airline, by Month")

sns.heatmap(data=flight_data, annot = True) # annot :     

plt.xlabel("Airline")
```

# Scatter Plots

```python
# Path of the file to read
insurance_filepath = "../data/Visualization_data/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
# add a regression line
sns.regplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
# Color-coded scatter plots
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'],hue=insurance_data['smoker'])
```

```python
# add two regression lines
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
# categorical scatter plot
sns.swarmplot(x=insurance_data['smoker'],y=insurance_data['charges'])
```
